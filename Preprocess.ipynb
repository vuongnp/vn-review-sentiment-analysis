```python
data = df.loc[:,['rating','content']]
```

```python
data['content']=data['content'].map(lambda x: str(x).lower())
data['content']=data['content'].map(lambda x: x.strip())
data.head()
```

```python
def words_convert(ori_s,dictionary):
    s =""
    words=ori_s.split(" ")
    for i in range(len(words)):
        if words[i] in dictionary.keys():
            words[i] = dictionary[words[i]]
        s=s+" "+words[i]
    return s.strip()
```

```python
viet_tat = {
    "sp" : "sản phẩm",
    "sx" : "sản xuất",
    "nv" : "nhân viên",
    "kcn" : "khu công nghiệp",
    "mk": "mình",
    "mik" : "mình",
    "e" : "em",
    "t" : "tôi",
    "b" : "bạn",
    "k" : "không",
    "r" : "rồi",
    "s" : "sao",
    "ko" : "không",
    "kg" : "không",
    "h" : "giờ",
    "bt" : "bình thường",
    "c" : " cái",
    "q" : " quyển",
    "j" : "gì",
    "dk" : "được",
    "đk" : "được",
    "đc" : "được",
    "hn" :"hà nội",
    "tphcm":"thành phố hồ chí minh",
    "hcm":"thành phố hồ chí minh",
    "nsx":"nhà sản xuất",
    "ok": "ô kê",
    "oki":"ô kê",
    "okie":"ô kê",
    "oke":"ô kê",
    "vs":"với"
}
```

```python
data['content']=data['content'].map(lambda ori_s: words_convert(ori_s,viet_tat))
data[20:30]
```

```python
englishs={
    "shop" : "cửa hàng",
    "sale" : "giảm giá",
    "info" : "thông tin",
    "test" : "kiểm tra",
    "check" : "kiểm tra",
    "good" : "tốt",
    "book" : "sách",
    "combo" : "gói",
    "full" : "đủ",
    "size" : "kích thước",
    "voucher": "phiếu quà tặng",
    "tks" : "cảm ơn",
    "thanks": "cảm ơn",
    "nice":"tuyệt",
    "thank": "cảm ơn",
    "online":"trực tuyến"
}
```

```python
data['content']=data['content'].map(lambda ori_s: words_convert(ori_s,englishs))
data[20:30]
```

```python
# kiểm tra từ có hai ký tự liền nhau giống nhau
def same_characters(w):
    for i in range(len(w)-1):
        if w[i]==w[i+1]:
            return True
    return False
def set_null_sentences(s):
    words=s.split(" ")
    for w in words:
        if same_characters(w):
            return np.nan
    return s
```

```python
data['content']=data['content'].map(lambda s: set_null_sentences(s))
data[20:30]
```

```python
def hasNumbers(inputString):
  return bool(re.search(r'\d', inputString))

def remove_digits(s):
  if hasNumbers(s):
    return np.nan
  return s
```

```python
data['content']=data['content'].map(lambda s: remove_digits(str(s)))
data[20:30]
```

```python
data.loc[data['rating'] <3, 'rating'] = 0
data.loc[data['rating'] == 3, 'rating'] = 1
data.loc[data['rating'] >3, 'rating'] = 2
data.head()
```

```python
data=data.dropna()
```

```python
neg = data[data['rating']==0][:1200]
neu = data[data['rating']==1][:1200]
pos = data[data['rating']==2][:1200]
print(neg.head())
print(neu.head())
print(pos.head())
```

```python
data = (neg.append(neu)).append(pos)
data.reset_index(inplace=True, drop=True)
```

```python
from vncorenlp import VnCoreNLP
rdrsegmenter = VnCoreNLP("vncorenlp/VnCoreNLP-1.1.1.jar", annotators="wseg", max_heap_size='-Xmx500m') 
```

```python
def tokenizer_data(rdrsegmenter, text):
  text = rdrsegmenter.tokenize(text)
  text = ' '.join([' '.join(x) for x in text])
  return text
```

```python
data['content']=data['content'].map(lambda s: tokenizer_data(rdrsegmenter, str(s)))
data[20:30]
```
