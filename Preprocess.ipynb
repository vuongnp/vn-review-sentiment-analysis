```python
def words_convert(ori_s,dictionary):
    s =""
    words=ori_s.split(" ")
    for i in range(len(words)):
        if words[i] in dictionary.keys():
            words[i] = dictionary[words[i]]
        s=s+" "+words[i]
    return s.strip()
```

```python
viet_tat = {
    "sp" : "sản_phẩm",
    "sx" : "sản_xuất",
    "nv" : "nhân_viên",
    "kcn" : "khu_công_nghiệp",
    "mk": "mình",
    "mik" : "mình",
    "e" : "em",
    "t" : "tôi",
    "b" : "bạn",
    "k" : "không",
    "r" : "rồi",
    "s" : "sao",
    "ko" : "không",
    "kg" : "không",
    "h" : "giờ",
    # "bt" : "bình thường",
    # "c" : " cái",
    # "q" : " quyển",
    "j" : "gì",
    "dk" : "được",
    "đk" : "được",
    "đc" : "được",
    # "hn" :"hà_nội",
    # "tphcm":"thành phố hồ chí minh",
    # "hcm":"thành phố hồ chí minh",
    "nsx":"nhà_sản_xuất",
    "ok": "ô_kê",
    "oki":"ô_kê",
    "okie":"ô_kê",
    "oke":"ô_kê",
    "vs":"với"
}
```
