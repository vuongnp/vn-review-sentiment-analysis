# Xử lý các từ viết tắt và từ tiếng Anh

```python
def words_convert(ori_s,dictionary):
    s =""
    words=ori_s.split(" ")
    for i in range(len(words)):
        if words[i] in dictionary.keys():
            words[i] = dictionary[words[i]]
        s=s+" "+words[i]
    return s.strip()
```

# Loại bỏ các câu chứa từ không đúng chính tả

```python
# kiểm tra từ có hai ký tự liền nhau giống nhau
def same_characters(w):
    for i in range(len(w)-1):
        if w[i]==w[i+1]:
            return True
    return False
def set_null_sentences(s):
    words=s.split(" ")
    for w in words:
        if same_characters(w):
            return np.nan
    return s
```

```python
data['content']=data['content'].map(lambda s: set_null_sentences(s))
data[20:30]
```

```python
# data.drop(data[data['content'].isna()].index, inplace = True)
data=data.dropna()
data[10:20]
```
