```python
import re
```

```python
def remove_emojis(text):
    emoj = re.compile("["
    u"\U0001F600-\U0001F64F"  # emoticons
    u"\U0001F300-\U0001F5FF"  # symbols & pictographs
    u"\U0001F680-\U0001F6FF"  # transport & map symbols
    u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
    u"\U00002500-\U00002BEF"  # chinese char
    u"\U00002702-\U000027B0"
    u"\U00002702-\U000027B0"
    u"\U000024C2-\U0001F251"
    u"\U0001f926-\U0001f937"
    u"\U00010000-\U0010ffff"
    u"\u2640-\u2642"
    u"\u2600-\u2B55"
    u"\u200d"
    u"\u23cf"
    u"\u23e9"
    u"\u231a"
    u"\ufe0f"  # dingbats
    u"\u3030"
                  "]+", re.UNICODE)
    
    return emoj.sub(r'', text)
    
```

```python
def words_convert(ori_s,dictionary):
    s =""
    words=ori_s.split(" ")
    for i in range(len(words)):
        if words[i] in dictionary.keys():
            words[i] = dictionary[words[i]]
        s=s+" "+words[i]
    return s.strip()
```

```python
def process(s):
  s = str(s).lower()
  s=remove_emojis(s)
  s=words_convert(s,viet_tat)
  s=words_convert(s,englishs)
  return s
```
